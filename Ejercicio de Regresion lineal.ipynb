```python
# separar caracteristicas de la etiquetas
X = datos_na.drop(['median_house_value'], axis=1)
y = datos_na['median_house_value']
```

```python
# Separar los datos en 2 partes: conjunto de entrenamiento y de pruebas
from sklearn.model_selection import train_test_split

X_ent, X_pruebas, y_ent, y_pruebas = train_test_split(X,y, test_size=0.2)
```

```python
y_pruebas.shape
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

```python
modelo.fit(X_ent, y_ent)
```

```python
predicciones = modelo.predict(X_pruebas)
```

```python
comparativa = {'prediccion': predicciones, 'valor real': y_pruebas }
pd.DataFrame(comparativa)
```

```python
# sobreajuste overfitting
print(modelo.score(X_ent, y_ent))
print(modelo.score(X_pruebas, y_pruebas))
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_pruebas, predicciones)
```
